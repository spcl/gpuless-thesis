```python
import pandas as pd
df = pd.read_csv('gpuless_local.csv')
print(df)
df.columns=["benchmark", "native_gpu", "native_gpu_total", "native_total", "gpuless_gpu", "gpuless_gpu_total", "gpuless_total"]
print(df)
```

```python
df['native_gpu'] = df['native_gpu'] / df['native_total'] * 100.0
df['native_gpu_total'] = df['native_gpu_total'] / df['native_total'] * 100.0
df['gpuless_gpu'] = df['gpuless_gpu'] / df['gpuless_total'] * 100.0
df['gpuless_gpu_total'] = df['gpuless_gpu_total'] / df['gpuless_total'] * 100.0
```

```python
cols = df.columns.tolist()
cols = [ cols[0], cols[3], cols[2], cols[1], cols[6], cols[5], cols[4] ]
df = df[cols]
final_df = df.groupby(['benchmark']).mean().round(2)
final_df
```
